```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df["District_Name"] = label_encoder.fit_transform(df["District_Name"])
```

```python
label_encoder = LabelEncoder()
df["Soil_color"] = label_encoder.fit_transform(df["Soil_color"])
```

```python
label_encoder = LabelEncoder()
df["Crop"] = label_encoder.fit_transform(df["Crop"])
```

```python
label_encoder = LabelEncoder()
df["Fertilizer"] = label_encoder.fit_transform(df["Fertilizer"])
```

```python
df.drop(columns=["Link"],inplace=True)
```

```python
df.head()
```

```python
x = df.drop(columns=["Fertilizer_Usage"])
y = df["Fertilizer_Usage"]
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)
```

```python
df["Soil_color"].unique()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
ypred = model.predict(xtest)
```

```python
# Step 8: Evaluate the model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(ytest, ypred)
mse = mean_squared_error(ytest, ypred)
r2 = r2_score(ytest, ypred)

# Print results
print("Mean Absolute Error:", mae) #errors
print("Mean Squared Error:", mse) #error
print("R-squared Score:", r2) #accuracy
```
